```python
# 현재 페이지 높이 구하기
page = driver.execute_script('return document.body.scrollHeight')

while True:

# 현재 페이지에서 스크롤 끝까지 내리기
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
    # 페이지에서 해당 태그가 로딩될때까지 기다림
    time.sleep(1.5)
    WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME, "Card_Card__WdaEk")))

    try:
        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME, "Card_Card__WdaEk")))
        # 스크롤 내린 후 페이지 높이 구하기
        new_page = driver.execute_script('return document.body.scrollHeight')
        if new_page == page:
            break
        else: page = new_page
    except:
        print('에러발생')
        break
```

```python
url = []
# 공고문에 해당하는 태그 선택
li = driver.find_elements(By.CLASS_NAME,'Card_Card__WdaEk')
for i in li:
    a = i.find_element(By.TAG_NAME,"a") # a 태그 가져오기
    href = a.get_attribute('href') # 링크 가져오기
    url.append(href)
```
